```python
import pandas as pd
import numpy as np
import random
import pickle
```

```python
df = pd.read_pickle('data/ingredients_cleaned.pickle')
```

```python
df['ingredients'].fillna('',inplace=True)
df
```

```python
def flatten(t):
    return [item for sublist in t for item in sublist]
from collections import Counter

counted = Counter(flatten(df['ingredients'].swifter.apply(lambda x: x.split(' ')).tolist()))
counted.most_common()
```

```python
from wordcloud import WordCloud
from PIL import Image

lemon_mask = np.array(Image.open("lemon_mask.png"))
wc_text = ' '.join(df['ingredients'].iloc[:100000].tolist())
wc = WordCloud(background_color="white", max_words=2000,mask=lemon_mask)
wc.generate(wc_text)
wc.to_file('test_2.png')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(lowercase=False,min_df=100)
doc_word = tfidf.fit_transform(df['ingredients'])
```

```python
with open('tfidf.pickle', 'wb') as f:
    pickle.dump(tfidf, f) 
```

```python
tfidf.transform(['lemon juice'])
```

```python
from sklearn.decomposition import TruncatedSVD

svd_model = TruncatedSVD(400)
doc_topic = svd_model.fit_transform(doc_word)

    
# svd_model.explained_variance_ratio_
```

```python
with open('doc_topic.pickle', 'wb') as f:
    pickle.dump(doc_topic, f) 
```

```python
with open('svd_model.pickle', 'wb') as f:
    pickle.dump(svd_model, f) 
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
        
display_topics(svd_model, tfidf.get_feature_names(), 5)
```
