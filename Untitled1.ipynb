```python
import pandas as pd
```

```python
df = pd.read_json('data/recipe1m.json')[['id','title','instructions','ingredients','quantity','unit','url']]
df['instructions'] = df['instructions'].apply(lambda x: [y['text'] for y in x])
df['ingredients'] = df['ingredients'].apply(lambda x: [y['text'] for y in x])
df['quantity'] = df['quantity'].apply(lambda x: [y['text'] for y in x])
df['unit'] = df['unit'].apply(lambda x: [y['text'] for y in x])
df
```

```python
def flatten(t):
    return [item for sublist in t for item in sublist]

set(flatten(df['unit'].tolist()))
```
